```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
################ Acurracy
ada020 = np.load('./results/evaluation_adaHessian020LR.npy')
adam001 = np.load('./results/evaluation_adam001LR.npy')
adagrad010 = np.load('./results/evaluation_adagrad010LR.npy')
sgd010 = np.load('./results/evaluation_sgd010LR.npy')
our02D015 = np.load('./results/evaluation_ourOptimizer2dir015LR.npy')

################ Losses
loss_ada020 = np.load('./results/losses_adaHessian020LR.npy')
loss_our015 = np.load('./results/losses_ourOptimizer2dir015LR.npy')
loss_adagrad010 = np.load('./results/losses_adagrad010LR.npy')
losses_adam001 = np.load('./results/losses_adam001LR.npy')
losses_sgd010 = np.load('./results/losses_sgd010LR.npy')
```

```python
################ Initial loss before optimization = 1/10
################ Initial cost before optimization = 3.7231
zero_acc = 10
zero_loss = 3.7231

loss_our = [zero_loss]
loss_ada = [zero_loss]
loss_adam = [zero_loss]
loss_adagrad = [zero_loss]
loss_sgd = [zero_loss]


for i in range(30):
    loss_our.append(loss_our015[i*125,1])
    loss_ada.append(loss_ada020[i*125,1])
    loss_adam.append(losses_adam001[i*125,1])
    loss_adagrad.append(loss_adagrad010[i*125,1])
    loss_sgd.append(losses_sgd010[i*125,1])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,6))


ax[0].plot(np.append(zero_acc,our02D015[:,0]))
ax[0].plot(np.append(zero_acc,ada020[:,0]))
ax[0].plot(np.append(zero_acc,adam001[:,0]))
ax[0].plot(np.append(zero_acc,adagrad010[:,0]))
ax[0].plot(np.append(zero_acc,sgd010[:,0]))


ax[1].plot(loss_our)
ax[1].plot(loss_ada)
ax[1].plot(loss_adam)
ax[1].plot(loss_adagrad)
ax[1].plot(loss_sgd)



legend = [
          "Our method",
          "Adahessian",
          "Adam",
          "Adagrad",
          "SGD",

             ]

plt.legend(legend,fontsize=16)

ax[0].set_xlabel("Epoch",fontsize=16)
ax[0].set_ylabel("Test accuracy",fontsize=16)
ax[1].set_ylabel("Training loss",fontsize=16)
ax[1].set_xlabel("Epoch",fontsize=16)

ax[0].tick_params(axis="x", labelsize=14) 
ax[0].tick_params(axis="y", labelsize=14) 
ax[1].tick_params(axis="x", labelsize=14) 
ax[1].tick_params(axis="y", labelsize=14) 

plt.savefig('experiment4.pdf')  
```
